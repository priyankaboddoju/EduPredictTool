Data Preprocessing and Feature Engineering

1. Data Cleaning: Handle Missing Values and Duplicates

```python
import pandas as pd

# Fill missing values for numerical columns with median
academic_numerical = academic.select_dtypes(include=['int64', 'float64'])
status_numerical = status.select_dtypes(include=['int64', 'float64'])
academic_detail_numerical = academic_detail.select_dtypes(include=['int64', 'float64'])

academic_numerical.fillna(academic_numerical.median(), inplace=True)
status_numerical.fillna(status_numerical.median(), inplace=True)
academic_detail_numerical.fillna(academic_detail_numerical.median(), inplace=True)

# Fill missing values for categorical columns with the mode
for col in academic.select_dtypes(include='object').columns:
    academic[col] = academic[col].fillna(academic[col].mode()[0])

for col in status.select_dtypes(include='object').columns:
    status[col] = status[col].fillna(status[col].mode()[0])

for col in academic_detail.select_dtypes(include='object').columns:
    academic_detail[col] = academic_detail[col].fillna(academic_detail[col].mode()[0])

# Remove duplicates from all datasets
academic.drop_duplicates(inplace=True)
status.drop_duplicates(inplace=True)
academic_detail.drop_duplicates(inplace=True)
```

```python
# Check if any missing values remain
missing_values_check = {
    "academic": academic.isnull().sum().sum(),
    "status": status.isnull().sum().sum(),
    "academic_detail": academic_detail.isnull().sum().sum()
}

# Check if duplicates remain
duplicates_check = {
    "academic": academic.duplicated().sum(),
    "status": status.duplicated().sum(),
    "academic_detail": academic_detail.duplicated().sum()
}

print("Missing Values After Cleaning:", missing_values_check)
print("Duplicate Rows After Cleaning:", duplicates_check)
```

2. Feature Engineering: Separate Categorical and Numerical Columns

```python
# Separate categorical and numerical columns
categorical_columns_academic = academic.select_dtypes(include='object').columns
numerical_columns_academic = academic.select_dtypes(include=['int64', 'float64']).columns

categorical_columns_status = status.select_dtypes(include='object').columns
numerical_columns_status = status.select_dtypes(include=['int64', 'float64']).columns

categorical_columns_academic_detail = academic_detail.select_dtypes(include='object').columns
numerical_columns_academic_detail = academic_detail.select_dtypes(include=['int64', 'float64']).columns
```

3. Encoding Categorical Variables: One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define the encoder for categorical columns
encoder_academic = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_columns_academic)
    ],
    remainder='passthrough'
)

encoder_status = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_columns_status)
    ],
    remainder='passthrough'
)

encoder_academic_detail = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_columns_academic_detail)
    ],
    remainder='passthrough'
)

# Apply encoding to datasets
academic_encoded = encoder_academic.fit_transform(academic)
status_encoded = encoder_status.fit_transform(status)
academic_detail_encoded = encoder_academic_detail.fit_transform(academic_detail)

# Check the shape after encoding
print("Academic encoded shape:", academic_encoded.shape)
print("Status encoded shape:", status_encoded.shape)
print("Academic detail encoded shape:", academic_detail_encoded.shape)
```

4. Scaling Numerical Variables: StandardScaler

```python
from sklearn.preprocessing import StandardScaler

# Define the scaler
scaler = StandardScaler()

# Apply scaling to numerical columns
academic_scaled = scaler.fit_transform(academic_numerical)
status_scaled = scaler.fit_transform(status_numerical)
academic_detail_scaled = scaler.fit_transform(academic_detail_numerical)

# Check the shapes after scaling
print("Academic scaled shape:", academic_scaled.shape)
print("Status scaled shape:", status_scaled.shape)
print("Academic detail scaled shape:", academic_detail_scaled.shape)
```

5. Combining the Encoded and Scaled Data

```python
import numpy as np

# Convert sparse matrices to dense arrays if needed
if not isinstance(academic_encoded, np.ndarray) and not isinstance(academic_encoded, pd.DataFrame):
    academic_encoded_dense = academic_encoded.toarray()
else:
    academic_encoded_dense = academic_encoded

if not isinstance(status_encoded, np.ndarray) and not isinstance(status_encoded, pd.DataFrame):
    status_encoded_dense = status_encoded.toarray()
else:
    status_encoded_dense = status_encoded

if not isinstance(academic_detail_encoded, np.ndarray) and not isinstance(academic_detail_encoded, pd.DataFrame):
    academic_detail_encoded_dense = academic_detail_encoded.toarray()
else:
    academic_detail_encoded_dense = academic_detail_encoded

# Ensure that both the scaled and encoded features are 2D arrays
academic_scaled_2d = academic_scaled if academic_scaled.ndim == 2 else academic_scaled.reshape(-1, 1)
status_scaled_2d = status_scaled if status_scaled.ndim == 2 else status_scaled.reshape(-1, 1)
academic_detail_scaled_2d = academic_detail_scaled if academic_detail_scaled.ndim == 2 else academic_detail_scaled.reshape(-1, 1)

# Get the feature names for encoded columns
encoded_feature_names_academic = encoder_academic.get_feature_names_out()
encoded_feature_names_status = encoder_status.get_feature_names_out()
encoded_feature_names_academic_detail = encoder_academic_detail.get_feature_names_out()


# Combine the scaled and encoded data
academic_final = pd.DataFrame(np.hstack([academic_scaled_2d, academic_encoded_dense]),
                              columns=np.concatenate([encoded_feature_names_academic, numerical_columns_academic]))
status_final = pd.DataFrame(np.hstack([status_scaled_2d, status_encoded_dense]),
                            columns=np.concatenate([encoded_feature_names_status, numerical_columns_status]))
academic_detail_final = pd.DataFrame(np.hstack([academic_detail_scaled_2d, academic_detail_encoded_dense]),
                                     columns=np.concatenate([encoded_feature_names_academic_detail, numerical_columns_academic_detail]))


# Check the final dataset shapes
print("Final Academic dataset shape:", academic_final.shape)
print("Final Status dataset shape:", status_final.shape)
print("Final Academic Detail dataset shape:", academic_detail_final.shape)
```

Step 5.4: Training the Random Forest Regressor model
